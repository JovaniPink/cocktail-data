```python
from sqlalchemy import create_engine

# Postgres username, password, and database name
# FInding the connection info, in the terminal jump to psql and run \conninfo
POSTGRES_ADDRESS = 'localhost' ## INSERT YOUR DB ADDRESS IS LOCALHOST
POSTGRES_PORT = '5432' ## PORT NUMBER
POSTGRES_USERNAME = 'postgres'     ## CHANGE THIS TO YOUR POSTGRES USERNAME
POSTGRES_PASSWORD = 'postgres' ## CHANGE THIS TO YOUR POSTGRES PASSWORD
POSTGRES_DBNAME = 'deverental' ## CHANGE THIS TO YOUR DATABASE NAME

# A long string that contains the necessary Postgres login information
postgres_str = ('postgresql://{username}:{password}@{ipaddress}:{port}/{dbname}'
                .format(username=POSTGRES_USERNAME, 
                        password=POSTGRES_PASSWORD,
                        ipaddress=POSTGRES_ADDRESS,
                        port=POSTGRES_PORT,
                        dbname=POSTGRES_DBNAME))

# Create the connection
cnx = create_engine(postgres_str)
```

```python
pd.read_sql_query('''SELECT * FROM customer LIMIT 5;''', cnx)
```

```python
pd.read_sql_query('''SELECT first_name FROM customer''', cnx)
```

```python
first_name_jamie = pd.read_sql_query(
'''SELECT
	first_name
FROM
	customer
WHERE
	first_name = 'Jamie'
''', cnx)
print(first_name_jamie)
```
